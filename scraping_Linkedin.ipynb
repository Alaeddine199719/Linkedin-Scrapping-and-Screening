```python
#new selectors of personnel informations
def Scraping_Data(link):
    soup=HTML_Sraping(link)
    time.sleep(10)
    final_liste=[]
    informations=dict()
    informations["link"]=link
    try:
        name_div= soup.find('div',{'class':'pv-text-details__left-panel mr5'})
        fullName = name_div.find('h1',{'class':'text-heading-xlarge inline t-24 v-align-middle break-words'}).get_text().strip()
        #print("fullName: ",fullName)
        informations["fullName"]=fullName
    except:
        print('No fullName')
        pass         
    #Personal info_Location  
    try:
        location_div= soup.find('div',{'class':'pv-text-details__left-panel mr5'})
        location= location_div.find('div',{'class':'pb2'}).find('span',{'class':'text-body-small inline t-black--light break-words'}).get_text().strip()
        #print("Location: ",location)
        informations["location"]=location
    except:
        print('No Location')
        pass 
    #Personal info_Profile_title
    try:
        
        profile_div= soup.find('div',{'class':'pv-text-details__left-panel mr5'})
        profile_title= location_div.find('div',{'class':'text-body-medium break-words'}).get_text().strip()
        #print("Profile Title: ",profile_title)
        informations["profile_title"]=profile_title
    except:
        print('No Profile Title')
        pass 
    
    #Personal info_ConnectionsCount
    try:
        name_loc = soup.find('ul',{'class':'pv-top-card--list pv-top-card--list-bullet display-flex pb1'})
        connection = name_loc.find('li',{'class':'text-body-small'})
        connectionsCount = connection.find('span',{'class':'t-black--light'}).get_text().strip()
        #print("Connections: ",connectionsCount)
        informations["connectionsCount"]=connectionsCount
    except:
        print('no Connections')
        
        pass 
    
    #Experience
    try:
        exp_section = soup.find('section', {'id': 'experience-section'}).find('ul')
        div_tag = exp_section.find_all('div',{'class': 'display-flex flex-column full-width'})
        job_title=[]
        company_name=[]
        joining_date=[]
        exp_duree=[]
        for i in range(len(div_tag)):
            a_tag = div_tag[i].find('a')
            job_title.append(a_tag.find('h3').get_text().strip())
            company_name.append(a_tag.find_all('p')[1].get_text().strip())
            joining_date.append(a_tag.find_all('h4')[0].find_all('span')[1].get_text().strip())
            exp_duree.append(a_tag.find_all('h4')[1].find_all('span')[1].get_text().strip())
            informations["job_title"]=job_title
            informations["company_name"]=company_name
            informations["joining_date"]=joining_date
            informations["exp_duree"]=exp_duree
    except:
        print('No Experiences')
        pass
    
    #Education: college_name
    try:
        edu_section = soup.find('section', {'id': 'education-section'}).find('ul')
        div_tag_edu = edu_section.find_all('div',{'class': 'display-flex flex-column full-width'})
        college_name=[]
        for i in range(len(div_tag_edu)):
            college_name.append(div_tag_edu[i].find('h3').get_text().strip())
            #print("college name: ",college_name)
            informations["college_name"]=college_name
    except:
        print('no Educations')
        pass
    
    #Education( degree_name)
    try:
        edu_section = soup.find('section', {'id': 'education-section'}).find('ul')
        div_tag_edu = edu_section.find_all('div',{'class': 'display-flex flex-column full-width'})
        degree_name=[]
        for i in range(len(div_tag_edu)):
            degree_name.append(div_tag_edu[i].find('p', {'class': 'pv-entity__secondary-title pv-entity__degree-name t-14 t-black t-normal'}).find_all('span')[1].get_text().strip())
            #print("degree name: ",degree_name)
            informations["degree_name"]=degree_name
    except:
        print('no data degree name')
        pass
    #Education(stream)
    try:
        edu_section = soup.find('section', {'id': 'education-section'}).find('ul')
        div_tag_edu = edu_section.find_all('div',{'class': 'display-flex flex-column full-width'})
        stream =[]
        for i in range(len(div_tag_edu)):
            stream.append(edu.find('p', {'class': 'pv-entity__secondary-title pv-entity__fos t-14 t-black t-normal'}).find_all('span')[1].get_text().strip())
            print("stream: ",stream)
            informations["stream"]=stream
    except:
        print('no data stream')
        pass
    #Education(degree_year)
    try:
        edu_section = soup.find('section', {'id': 'education-section'}).find('ul')
        div_tag_edu = edu_section.find_all('div',{'class': 'display-flex flex-column full-width'})
        degree_year=[]
        for i in range(len(div_tag_edu)):
            degree_year.append(div_tag_edu[i].find('p', {'class': 'pv-entity__dates t-14 t-black--light t-normal'}).find_all('span')[1].get_text().strip())
            informations["degree_year"]=degree_year
    except:
        print('no data degree year')
        pass 
    
    #competences
    try:
        skills=[]
        for i in range(len(soup.find_all('div', {'class':'pv-skill-category-entity__skill-wrapper'}))):
            skills.append(soup.find_all('div', {'class':'pv-skill-category-entity__skill-wrapper'})[i].find('p',{'class':'pv-skill-category-entity__name tooltip-container'}).get_text().strip())
            informations["skills"]=skills
    except:
        print('No skills')
        pass
    
    #Recommendations  
    try:
        en=[]
        for i in range(len(soup.find_all('span', {'class':'custom-highlight t-bold'}))):
            en.append(soup.find_all('span', {'class':'custom-highlight t-bold'})[i].get_text().strip())
            #print("endorsements: ",en) 
            informations["endorsements"]=en
    except:
        print('no endorsements')
        pass  
    
    #lANGAGEs
    try:
        Accomp_section = soup.find('section', {'class': 'pv-profile-section pv-accomplishments-block languages ember-view'})  
        language=[]
        for i in range(len(Accomp_section.find('ul').find_all('li'))):
            language.append(Accomp_section.find('ul').find_all('li')[i].get_text().strip())
            #print("language: ",language)
            informations["language"]=language
    except:
        print('No Languages')
        pass
    #projects
    try:
        projets_section=soup.find('section', {'class': 'pv-profile-section pv-accomplishments-block projects pv-accomplishments-block--last ember-view'})        
        pp=[]
        projets_section.find_all('li', {'class':"pv-accomplishments-block__summary-list-item"})
        for i in range(len(projets_section.find_all('li', {'class':"pv-accomplishments-block__summary-list-item"}))):
            pp.append(projets_section.find_all('li', {'class':"pv-accomplishments-block__summary-list-item"})[i].get_text().strip())
            informations["Projects"]=pp
    except:
        print('no project')
        pass
    #Nomber of projects
    try:
        projets_section=soup.find('section', {'class': 'pv-profile-section pv-accomplishments-block projects pv-accomplishments-block--last ember-view'})        
        for i in projets_section.find("h3").find('span', {'class':"visually-hidden"}).get_text().strip():        
            try:
                (int(i))
                informations["Project numbers"]=int(i)
            except:
                pass
    except:
        print('aucun projet exist')
        pass
    #Publications
    try:
        publication_section = soup.find('div', {'id':'publications-expandable-content'})
        p=[]
        nb=0
        for i in range(len(soup.find_all('div', {'id':'publications-expandable-content'}))):
            p.append(soup.find_all('div', {'id':'publications-expandable-content'})[i].find('li').get_text().strip())
            nb=nb+1
            informations["Publications"]=p
            informations["Pub numbers"]=nb
    except:
        print('no Publication')
        pass 
    #Interests
    try:
        interests_section = soup.find('section', {'class':"pv-profile-section pv-interests-section artdeco-card mt4 p5 ember-view"})
        interests_name=[]
        for i in range(len(interests_section.find_all('div', {'class':"pv-entity__summary-info ember-view"}))):
            inter=interests_section.find_all('div', {'class':"pv-entity__summary-info ember-view"})[i]
            interests_name.append(inter.find('span').get_text().strip())
            informations["interests"]=interests_name
    except:
        print('no interests')
        pass
    #Certifications
    try:
        certifications=[]
        for i in range(len(soup.find_all('h3', {'class':"t-16 t-bold"}))):
            certif=soup.find_all('h3', {'class':"t-16 t-bold"})[i].get_text().strip()
            certifications.append(certif)
            informations["certifications"]=certifications
    except:
        print('no certifications')
        pass
    
    time.sleep(20)
    final_liste.append(informations)
    print(informations)
   
    return(final_liste)
```
